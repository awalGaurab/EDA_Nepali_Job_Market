```python
import os
import csv
import time
import requests
from bs4 import BeautifulSoup

# print(soup.prettify())
```

```python
for page_number in range(1,74):
    URL = "https://merojob.com/search/?page={}".format(page_number)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")

    job_list = []
    company_list = []
    location_list = []
    skill_list = []
    
    # To get job titles for one page   
    job_titles = soup.find_all('h1',itemprop="title")
    for job_title in job_titles:
        job_list.append(job_title.text.strip())

    # To read company names    
    companies = soup.find_all('a',class_='text-dark')
    for company in companies:
        company_list.append(company.text.strip())

    # To read all addresses of companies at a single time
    locations = soup.find_all('span',itemprop='addressLocality')
    for location in locations:
        location_list.append(location.text.strip())

    # skill_list contains list of list with all skills for each page
    skills = (soup.find_all('span', itemprop="skills"))
    for skill in skills:
        skill_text = skill.text.split("\n")
        skill_list.append(skill_text)       

    for eachlist in skill_list:
            eachlist.remove("")     
    for eachlist in skill_list:
            eachlist.remove("")    
#     print(skill_list)
    
    for i in range(6):
        filename="merojob.csv"
        headers = ['Job title','Company name','Location','Skills']
        file_exists = os.path.isfile(filename)
        
        try:
            job = job_list[i]
        except Exception as e:
            print("Exception is ",e)
            print("Job title, page number and card number",page_number,i )
            job = " "
        
        try:
            company = company_list[i]
            
        except Exception as e:
            print("Exception at company name ",e)
            print("Company name, page number and card number",page_number,i )
            company = " "
            
        try:
            location = location_list[i]
            
        except Exception as e:
            print("Exception at Location/Address ",e)
            print("Location/Address, page number and card number",page_number,i )
            location = " "
        
        try:
            skill = (','.join(skill_list[i]))
        except Exception as e:
            print("Exception at skills ",e)
            print("Skills, page number and card number",page_number,i )
            skill = ""
        
        each_row = [job,company,location,skill]

        with open(filename, mode='a', newline='', encoding='utf-8') as file:
            file_writer = csv.writer(file, delimiter=',')
            if not file_exists:
                file_writer.writerow(headers)
            file_writer.writerow(each_row)
#             print("Page number and each row data",page_number,each_row)
    
    
    time.sleep(2)
```
